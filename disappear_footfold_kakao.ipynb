```python
import copy

def pos_ck(h,w):
    global bh, bw
    clist = []
    if h != 0:
        clist.append((h-1,w))
    if h != bh-1:
        clist.append((h+1,w))
    if w != 0:
        clist.append((h,w-1))
    if w != bw-1:
        clist.append((h,w+1))
        
    return clist

def win_ck(can):
    win = False
    Min = 26
    Max = 0
    
    for c in can:
        if c[0]==0:
            win = True
            break
    
    if win: #승
        for c in can:
            if c[0]==0:
                if c[1] < Min:
                    Min = c[1]
        return (1,Min)
    else: #패
        for c in can:
            if c[1] > Max:
                Max = c[1]
        return (0,Max)

def move_a(aloc,bloc,move,board):
    if board[aloc[0]][aloc[1]] == 0:
        return (0,move)
    clist = pos_ck(aloc[0],aloc[1])
            
    can = []
    for c in clist:
        H,W = c[0],c[1]
        if board[H][W]:
            board[aloc[0]][aloc[1]] = 0
            ret = move_b([H,W],bloc,move+1,board)
            board[aloc[0]][aloc[1]] = 1
            can.append(ret)
    if can == []:
        return (0,move)
    else:
        return win_ck(can)
    
def move_b(aloc,bloc,move,board):
    if board[bloc[0]][bloc[1]] == 0:
        return (0,move)
    clist = pos_ck(bloc[0],bloc[1])
    
    can = []
    for c in clist:
        H,W = c[0],c[1]
        if board[H][W]:
            board[bloc[0]][bloc[1]] = 0
            ret = move_a(aloc,[H,W],move+1,board)
            board[bloc[0]][bloc[1]] = 1
            can.append(ret)
    if can == []:
        return (0,move)
    else:
        return win_ck(can)

bh,bw = 0,0
def solution(board, aloc, bloc):
    global bh,bw
    
    bh , bw = len(board), len(board[0])
    ans = move_a(aloc,bloc,0,board)
    answer = ans[1]
    return answer
```
